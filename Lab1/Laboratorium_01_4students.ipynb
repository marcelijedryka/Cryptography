### Szyfr Cezara
--------------
Szyfr Cezara oczywiście nie zawdzięcza swego istnienia Juliuszowi Cezarowi. Został przezeń jedynie wspomniany w pamiętnikach o wojnach galijskich. Prawdopodobnie była to dobrze znana metoda w tym czasie.

```python
# Alfabet jawny i szyfrowy
alphabet = ['a','b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p',
            'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

alphabet_size = 26

plaintext = "Zorro fly zealotry zipper"

key = 15 #przesunięcie permutacji czyli klucz szyfrowania

print("Tekst jawny: ", plaintext)

def cezar_coder(plaintext):

    ciphertext = ""

    for text in plaintext.lower().split():
        for char in text:
            ciphertext = ciphertext + alphabet[(alphabet.index(char) + key) % alphabet_size]
    
    return ciphertext

ciphertext = cezar_coder(plaintext)
print("Szyfrogram: ",  ciphertext, "(spacje usunięte)")

def cezar_decoder(ciphertext):
    decodedtext = ""
    for text in ciphertext:
        for char in text:
            decodedtext = decodedtext + alphabet[(alphabet.index(char)- key) % alphabet_size]
    
    return decodedtext

decodedtext = cezar_decoder(ciphertext)
print("Odszyfrowana wiadomość: ",  decodedtext,  "(spacje usunięte)")
```

```python
# lepszy Cezar
import numpy as np
plaintext = "Zorro fly zealotry zipper"


permutation = np.random.permutation(alphabet).tolist()

def better_cezar_coder(plaintext , permutation):
    cipher = ""
    for word in plaintext.lower().split():
        for letter in word:
            cipher += alphabet[permutation.index(letter)]
    
    return cipher

cipher = better_cezar_coder(plaintext , permutation)

print("Coded text: " + cipher)

def better_cezar_decoder(cipher , permutation):
    decoded = ""

    for letter in cipher:
        decoded += permutation[alphabet.index(letter)]
    
    return decoded

decoded = better_cezar_decoder(cipher,permutation)
    
print("Decoded text: " + decoded)
```

```python
# policz częstość i wyświetl rozkłady

def analize_string_frequency(s):
    n = len(s)
    letter_count = dict()

    for char in s.lower():
        if char not in letter_count.keys():
            letter_count[char] = 1
        else:
            letter_count[char] += 1

    freq = []

    for letter in alphabet:
        if letter in letter_count:
            freq.append(letter_count[letter] / n)
        else:
            freq.append(0)
        
    return freq




plaintext = "tobeornottobethatisthequestionwhethertisnoblerinthemindtosuffertheslingsandarrowsofoutrageousfortuneortotakearmsagainstaseaoftroublesandbyopposingendthem"

cezar_cipher = cezar_coder(plaintext)
better_cezar_cipher = better_cezar_coder(plaintext , permutation)

frequency = analize_string_frequency(plaintext)
lab = np.arange(26)
plt.figure(figsize=(15,9))
plt.bar(lab, frequency, alpha=0.5, edgecolor="green")
plt.xlabel('Alfabet')
plt.xticks(np.arange(26), alphabet)
plt.title("Przykład");
plt.ylabel('Częstość występowania')
plt.show()
```
